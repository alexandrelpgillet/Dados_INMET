```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Nome:GOIANIA
#Código Estação: A002
#Latitude: -16.64277777
#Longitude: -49.2202777
#Altitude:727.3
#Situação: Operante
#Data Inicial:2001-05-28
#Data Final: 2024-10-22
#Periodicidade da Medição:Horaria 


Data = pd.read_csv('E:/Dados_INMET/dados_A002_H_2001-05-28_2024-10-22.csv',sep=';')


Data.head()
```

```python
Data.dropna(inplace=True)
```

```python
Data.tail()
```

```python
Data["Data Medicao"] = pd.to_datetime(Data["Data Medicao"],dayfirst=True, errors='coerce')

Data = Data[Data["Data Medicao"].dt.year==2023]
```

```python
Data
```

```python

Temp = []
Data_media =[]
Data_desvio_padrao=[]
Data_primeiro_quartil=[]
Data_segundo_quartil=[]
Data_terceiro_quartil=[]
Data_mediana=[]
Data_intervalo_interquartil=[]

for i in range(1,13):
    
    
    aux = Data[Data["Data Medicao"].dt.month==i]
    Temp.append(aux["TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)"])
    
    
    Data_media.append(aux["TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)"].mean())
    Data_desvio_padrao.append(aux.std())
    Data_primeiro_quartil.append(aux.quantile(0.25))
    Data_segundo_quartil.append(aux.quantile(0.50))
    Data_terceiro_quartil.append(aux.quantile(0.75))
    Data_mediana.append(aux.median())
    Data_intervalo_interquartil.append(aux.quantile(0.75)-aux.quantile(0.25))
    

```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.plot(meses,Data_media, marker='o', linestyle='-', color='blue', label='Média Temperatura Mensal')

plt.title("Média Temperatura Mensal de Goiânia")

plt.xlabel("Meses")
plt.ylabel("Temperatura Mensal")
plt.grid(True)

plt.show()
```

```python
plt.boxplot(Temp,vert=True , patch_artist=True)
plt.xlabel('Meses')
plt.ylabel('Temperatura')
plt.title("Boxplot")
plt.grid(True)
plt.show()
```

```python
plt.bar(meses, Data_media , color='blue')
plt.xlabel('Meses')
plt.ylabel('Temperatura')
plt.grid(True)
plt.show()
```
